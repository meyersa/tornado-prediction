```python
from netCDF4 import Dataset

# Path to the NetCDF file
file_path = 'NUL_130915_183202_KTBW_476187s_U2.nc'

# Open the NetCDF file
with Dataset(file_path, 'r') as nc:
    # Print a summary of the file
    print("File Summary:")
    print(nc)

    # List all variables
    print("\nVariables:")
    for var_name in nc.variables.keys():
        var = nc.variables[var_name]
        print(f"{var_name}: {var.dimensions}, shape: {var.shape}, dtype: {var.dtype}")

    # List all dimensions
    print("\nDimensions:")
    for dim_name, dim in nc.dimensions.items():
        print(f"{dim_name}: size {len(dim)}")

    # List global attributes
    print("\nGlobal Attributes:")
    for attr_name in nc.ncattrs():
        print(f"{attr_name}: {getattr(nc, attr_name)}")
```
